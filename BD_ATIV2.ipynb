```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
pd.set_option('display.max_colwidth', None)

def extrair_informacoes_professores():
    url = "https://www.ifpb.edu.br/ppgti/programa/corpo-docente"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    title = soup.title.string
    print(title)

    informacoes_professores = []

    container_professores = soup.find('div', id='parent-fieldname-text')
    for tag_professor in container_professores.find_all('h4'):
        nome = tag_professor.text.strip()
        linha_pesquisa = tag_professor.find_next_sibling('p').get_text()
        url_lattes = tag_professor.find_next_sibling('p').find('a')['href']
        email = tag_professor.find_next_sibling('p').find('br')

        informacoes_professores.append({
            'Nome': nome,
            'Linha de Pesquisa': linha_pesquisa,
            'Lattes': url_lattes,
            'E-mail': email
        })

    return informacoes_professores

informacoes_professores = extrair_informacoes_professores()
# Criando o dataframe a partir do scraping
df_docentes = pd.DataFrame(informacoes_professores)
# Realizando a extração do email e linha de pesquisa usando expressões regulares
linha = r":\s*(.*?)\s*Currículo"
email = r"E-mail:\s*([^\s]+@[^\s]+)"
df_docentes['E-mail'] = df_docentes['Linha de Pesquisa'].str.extract(email)
df_docentes['Linha de Pesquisa'] = df_docentes['Linha de Pesquisa'].str.extract(linha)
df_filtrado = df_docentes.loc[df_docentes['Nome'] != '']
df_filtrado
```
